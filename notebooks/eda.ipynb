```python
%%sql
-- looking at 
SELECT *, s/c AS p_insured 
FROM (
    SELECT st, SUM(c) s, COUNT(1) c FROM (
        SELECT st, CASE WHEN hins1+hins2+hins3+hins4+hins5+hins6+hins7 > 0 THEN 1 ELSE 0 END c FROM test.silver.data.census
    ) GROUP BY st
) ORDER BY p_insured DESC
```
